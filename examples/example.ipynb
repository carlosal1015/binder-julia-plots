# Example

This example shows that having TeX dependencies and Julia packages preinstalled allows TeX to be used when creating plots.

Importing the package:

```julia
using PyPlot
```

Generating data:

```julia
x = range(0; length = 1000)
y = sin.(Ï€ .* x ./ 100);
```

Customizing plots:

```julia
# Get access to rcParams
rcP = PyPlot.PyDict(PyPlot.matplotlib."rcParams")

# Set DPI
rcP["savefig.dpi"] = 300
rcP["figure.dpi"] = 300

# Activate TeX support
rcP["text.usetex"] = true

# Activate languages support
rcP["text.latex.preamble"] = raw"\usepackage[main=english]{babel}"

# Switch to Computer Modern font set
rcP["mathtext.fontset"] = "cm"

# Set font sizes
rcP["font.size"] = 18
rcP["legend.fontsize"] = 12;
```

Creating a plot:

```julia
plot(x, y, color = "#425378")
title(raw"\textrm{A sample from} $\sin{\frac{\pi x}{100}}$", y = 1.02)
xlabel(raw"$x$")
ylabel(raw"$y$");
```
